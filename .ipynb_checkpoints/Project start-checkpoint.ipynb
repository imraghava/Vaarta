```python
import pandas as pd
import glob
from skimage import data, io

################################################################################

def getfiles(filenames , data_directory): #to extract images from all folders in a dictionary
    dir_files = {}
    for x in filenames:
        dir_files[x]=io.imread(data_directory+x)
    return dir_files

###############################################################################

user_list = ["user_3" , "user_4", "user_5" , "user_6"]
data_directory = "/Users/melon/Downloads/Dataset/"

#################################################################################
frame = pd.DataFrame() #creating empty data frame
list_ = [] 
for user in user_list:
    list_.append(pd.read_csv(data_directory+user+'/'+user+'_loc.csv',index_col=None,header=0)) 
    #Reading all csv files for each folder(user) and appending the data of csv files in the list
    
    
    
frame = pd.concat(list_) #in a way making a list a dataframe
frame['side']=frame['bottom_right_x']-frame['top_left_x'] # Adding new columns in data frame for rectangle (not required in our case)
frame['hand']=1 #Telling the hand is present in our image (again not reqired in our case)
#print(frame)
#print(frame.image.unique()) 

imageset = getfiles(frame.image.unique() , data_directory)

```
