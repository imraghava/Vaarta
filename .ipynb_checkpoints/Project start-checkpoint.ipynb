## Loading the images for my dataset, following same approach


```python
import pandas as pd 
from skimage import io
from skimage.feature import hog
from skimage.color import rgb2gray

def getfiles(filenames , data_directory):
    img_dict = {}
    for x in filenames:
        img_dict[x] = io.imread(data_directory +x)
    print(img_dict)
    return img_dict
############################################################

user_list = ["user1"]
data_directory = "/Users/melon/data/"
##############################################################
def imgDictionary(user_list , data_directory):
    df = pd.DataFrame()
    lst = []
    for user in user_list:
        lst.append(pd.read_csv(data_directory + user + "/" + user + ".csv" , header = 0 , names = ["image"] , index_col = None))

    print(lst)
    #df = pd.concat(lst)
    img_dict = getfiles(lst , data_directory)
    
    return img_dict

def convertToGrayToHog(image):
    gray = rgb2gray(image)
    return hog(gray)
    

```

```python
def load_data(user_list , data_directory , img_dict):
    X = []
    Y = []
    
    for key in img_dict:
        print(key)
        hogVector = convertToGrayToHog(img_dict[key])
        X.append(hogVector.tolist())
        Y.append(key.split("_")[1])
        if(key == "user1/hand1_4_dif_seg_2_cropped.png"):
            break
    return X, Y

img_dict = imgDictionary(user_list , data_directory)
```
